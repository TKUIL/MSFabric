```python
# variable defined in the pipeline
entities_info = eval(entities_info)

def process_entities(entities_info):
    for entity in entities_info:       
        try:
            entity_id = entity["entity_id"]
            file_path = entity["file_path"]
            
            file_processor = FileProcessor(entity_id=entity_id, file_path=file_path)
            file_processor.write_file_to_bronze()
            file_processor.write_file_to_silver()
        
        except Exception as e:
            print(f"Error processing entity_id {entity_id} with file {file_path}: {e}")

process_entities(entities_info)
```
